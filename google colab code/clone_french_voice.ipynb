```python
import subprocess
from TTS.api import TTS
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2", gpu=True, progress_bar=True)
# Path to your voice sample, replace if needed
speaker_wav_path = "example.wav"
```

```python
from google.colab import files
uploaded = files.upload()
for filename in uploaded.keys():
    print(filename)
```

```python
for filename in list(uploaded.keys()):
    with open(filename, "r") as f:
        text = f.read()
    tts.tts_to_file(text=text, file_path=f"{filename}_output.wav", speaker_wav=speaker_wav_path, language="fr")
    subprocess.run([
        "ffmpeg", "-i", f"{filename}_output.wav",
        "-filter:a", "atempo=0.8",
        f"{filename}_slow.wav"

    ])
    files.download(f"{filename}_slow.wav")
```
