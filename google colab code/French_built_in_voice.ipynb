```python
from TTS.api import TTS

tts = TTS(model_name="tts_models/fr/css10/vits", gpu=True)

text = "Bonjour, je suis une voix française générée par l'intelligence artificielle."
tts.tts_to_file(text=text, file_path="french_voice.wav")
```

```python
import subprocess
from google.colab import files
uploaded = files.upload()
for filename in uploaded.keys():
    print(filename)
```

```python
for filename in list(uploaded.keys()):
    with open(filename, "r") as f:
        text = f.read()
    tts.tts_to_file(text=text, file_path=f"{filename}_french_voice.wav")
    subprocess.run([
        "ffmpeg", "-i", f"{filename}_french_voice.wav",
        "-filter:a", "atempo=0.8",
        f"{filename}_model_2.wav"
    ])
    files.download(f"{filename}_model_2.wav")
```
